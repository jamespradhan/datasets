```python
import pandas as pd
import yfinance as yf
import datetime
```

```python
raw_data = pd.read_csv("raw/EQUITY_L.csv")

raw_data.head()
```

```python
raw_symbols = raw_data['SYMBOL'].to_list()

print(raw_symbols)
```

```python
symbols = []
names = []
sectors = []
market_caps = []
dates = []
errors = []
for symbol in raw_symbols:
    symbol = symbol + ".NS"
    try:
        ticker = yf.Ticker(symbol)
        info = ticker.info
        if info['market'] != 'in_market':
            print(f"{symbol} is from {info['market']}")
            continue
        
        name = info.get('longName') or info.get('shortName') or info.get('name') or None
        sector = info.get('sector') or None
        market_cap = info.get('marketCap') or None
        print(f"{symbol} - {name} - {sector} - {market_cap}")
        timestamp = info['firstTradeDateMilliseconds']
        date = datetime.datetime.fromtimestamp(timestamp / 1000, tz=datetime.timezone.utc)
    except:
        print(f"{symbol} ERROR !!!")
        errors.append(symbol)

    # print(f"{symbol} - {name} - {sector} - {market_cap} - {date}")

    symbols.append(symbol)
    names.append(name)
    sectors.append(sector)
    market_caps.append(market_cap)
    dates.append(date)

print("COMPLLETE !!!")
```

```python
df = pd.DataFrame({
    "symbol": symbols,
    "name": names,
    "sector": sectors,
    "market_cap": market_caps,
    "date_of_listing": dates, 
})
df.head()
```

```python
df.info()
```

```python
df.to_csv('data/nse-all-market.csv', index=False)
```
